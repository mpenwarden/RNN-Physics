### Double Pendulum:
Given: $l_1,l_2,m_1,m_2, \theta_1(0),\theta_2(0),g$ <br>
$x_1 = l_1 sin(\theta_1)$ <br>
$y_1 = -l_1 cos(\theta_1)$ <br> 
$x_2 = l_1 sin(\theta_1) + l_2 sin(\theta_2)$ <br>
$y_2 = -l_1 cos(\theta_1) - l_2 cos(\theta_2)$ <br>

$v_{x1} = l_1 cos(\theta_1)\dot{\theta_1}$ <br>
$v_{y1} = l_1 sin(\theta_1)\dot{\theta_1}$ <br> 
$v_{x2} = l_1 cos(\theta_1)\dot{\theta_1} + l_2 cos(\theta_2)\dot{\theta_2}$ <br>
$v_{y2} = l_1 sin(\theta_1)\dot{\theta_1} + l_2 sin(\theta_2)\dot{\theta_2}$ <br>

$v_1^2 = v_{x1}^2+v_{y1}^2 = l_1^2 \dot{\theta_1}^2$<br>
$v_2^2 = l_1^2 \dot{\theta_1}^2 + l_2^2 \dot{\theta_2}^2 + 2 l_1 l_2 \dot{\theta_1} \dot{\theta_2} cos(\theta_1-\theta_2) $<br>


### Lagrangian Mechanics:
$L = T - V$ <br>
$T = \frac{1}{2}m_1 v_1^2 + \frac{1}{2}m_2 v_2^2$ <br>
$V = m_1 g y_1 + m_2 g y_2$ <br>
$\Rightarrow L = \frac{1}{2}(m_1+m_2)l_1^2 \dot{\theta_1^2} + \frac{1}{2}m_2 l_2^2 \dot{\theta_2^2} + m_2 l_1 l_2 \dot{\theta_1} \dot{\theta_2} cos(\theta_1-\theta_2) + (m_1+m_2)gl_1 + m_2 g l_2 cos(\theta_2) $ <br>

Euler-Lagrange equation: $\frac{d}{dt}(\frac{\partial L}{\partial \dot{q}_i}) = \frac{\partial L}{\partial q_i}$ <br>
Let $q_i \in [\theta_1,\theta_2]$ <br>
Lagrange Equation of Motion (EoM) for $\theta_1$: <br>
$(m_1+m_2) l_1 \ddot{\theta_1} + m_2 l_2 \ddot{\theta_2} cos(\theta_1-\theta_2) - m_2 l_2 \dot{\theta_2} sin(\theta_1-\theta_2)(\dot{\theta_1}-\dot{\theta_2) = -g(m_1+m_2)sin(\theta_1) - m_2 l_2 \dot{\theta_1} \dot{\theta_2} sin(\theta_1 - \theta_2})$<br>

Lagrange EoM for $\theta_2$: <br>
$m_2 l_2 \ddot{\theta_2} + m_2 l_1 \ddot{\theta_1} cos(\theta_1-\theta_2) - m_2 l_1 \dot{\theta_1^2} sin(\theta_1-\theta_2) = -gm_2sin(\theta_2) $<br>


### Hamiltonian Mechanics
$H(p_i,q_i) = \sum_i p_i \dot{q_i} - L$ <br>
$p_i = \frac{\partial L}{\partial \dot{q_i}}$ <br>

(1) &nbsp;&nbsp;&nbsp;  $p_1 = \frac{\partial L}{\partial \dot{\theta_1}} = (m_1+m_2)l_1^2 \dot{\theta_1} + m_2 l_1 l_2 \dot{\theta_2} cos(\theta_1-\theta_2)$ <br>
(2) &nbsp;&nbsp;&nbsp;  $p_2 = \frac{\partial L}{\partial \dot{\theta_2}} = m_2l_2^2 \dot{\theta_2} + m_2 l_1 l_2 \dot{\theta_1} cos(\theta_1-\theta_2)$ <br>
$\Rightarrow \sum_i p_i \dot{q_i} = (m_1+m_2)l_1^2 \dot{\theta_1^2} + m_2l_2^2\dot{\theta_2^2} + 2 m_2l_1l_2 \dot{\theta_1} \dot{\theta_2} cos(\theta_1-\theta_2)$ <br>

$\Rightarrow H(p_i,q_i) = \frac{1}{2}(m_1+m_2)l_1^2\dot{\theta_1^2} + \frac{1}{2}m_2l_2^2\dot{\theta_2^2}+m_2l_1l_2\dot{\theta_1}\dot{\theta_2}cos(\theta_1-\theta_2)+gl_1(m_1+m_2)cos(\theta_1)-m_2gl_2cos(\theta_2)$ <br>

### Deriving Hamilton EoM's
Using (1) & (2) we solve for $\dot{\theta_1}$ & $\dot{\theta_2}$ <br>
(3) &nbsp;&nbsp;&nbsp; $\dot{\theta_1} = \frac{l_2p_1-l_1p_2cos(\theta_1-\theta_2)}{l_1^2l_2(m_1+m_2sin^2(\theta_1-\theta_2)} = \frac{\partial H}{\partial p_1}$ <br>
(4) &nbsp;&nbsp;&nbsp; $\dot{\theta_2} = \frac{l_1(m_1+m_2)p_2 - l_2m_2p_1cos(\theta_1-\theta_2)}{l_1l_2^2m_2(m_1+m_2sin^2(\theta_1-\theta_2))} = \frac{\partial H}{\partial p_2} $<br>

Substituting back into H: $\Rightarrow H = \frac{l_2^2 m_2 p_1^2 + l_1^2(m_1+m_2)p_2^2 - 2m_2l_1l_2p_1p_2cos(\theta_1-\theta_2)}{2l_1^2l_2^2m_2(m_1+sin^2(\theta_1-\theta_2)m_2)} - m_2gl_2cos(\theta_2) -(m_1+m_2)gl_1cos(\theta_1)$<br>
(5) &nbsp;&nbsp;&nbsp; $\dot{p_1} = -\frac{\partial H}{\partial \theta_1} = -(m_1+m_2)gl_1sin(\theta_1)-C_1+C_2$<br>
(6) &nbsp;&nbsp;&nbsp; $\dot{p_2} = -\frac{\partial H}{\partial \theta_2} =  -m_2gl_2sin(\theta_2)+C_1+C_2$<br>
where $C_1 = \frac{p_1p_2sin(\theta_1-\theta_2)}{l_1l_2(m_1+m_2sin^2(\theta_1-\theta_2)}$ <br>
and $C_2 = \frac{l_2^2 m_2 p_1^2 + l_1^2(m_1+m_2)p_2^2-l_1l_2m_2p_1p_2cos(\theta_1-\theta_2)}{2l_1^2l_2^2(m_1+m_2sin^2(\theta_1-\theta_2)^2}sin(2(\theta_1-\theta_2))$ <br>

Therefore (3),(4),(5),(6) are the Hamiltonian Equations of Motion

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy import integrate
import tensorflow as tf
```

```python
class DoublePendulum:
    def __init__(self,theta1,theta2,l1,l2,m1,m2,dt,t_end):
        self.theta1 = theta1
        self.theta2 = theta2
        self.p1 = 0
        self.p2 = 0
        self.dt = dt
        self.t_end = t_end
        self.g = 9.81
        self.l1 = l1
        self.l2 = l2
        self.m1 = m1
        self.m2 = m2
        
    # Hamiltonian system
    def f(self,t,r):
        theta1, theta2, p1, p2 = r
        l1 = self.l1
        l2 = self.l2
        m1 = self.m1
        m2 = self.m2
        g = self.g
        
        C1 = (p1*p2*np.sin(theta1-theta2))/(l1*l2*(m1+m2*np.sin(theta1-theta2)**2))
        C2 = ((l2**2 * m2 * p1**2 + l1**2 * (m1+m2)*p2**2 - l1*l2*m2*p1*p2*np.cos(theta1-theta2))/(2*l1**2 * l2**2 * (m1+m2*np.sin(theta1-theta2)**2)**2))*(np.sin(2*(theta1-theta2)))

        theta1_dot = (l2*p1-l1*p2*np.cos(theta1-theta2))/(l1**2 * l2 * (m1+m2*np.sin(theta1-theta2)**2))
        theta2_dot = (l1*(m1+m2)*p2 - l2*m2*p1*np.cos(theta1-theta2))/(l1*l2**2 * m2 *(m1+m2*np.sin(theta1-theta2)**2)) 
        p1_dot = -(m1+m2)*g*l1*np.sin(theta1)-C1+C2
        p2_dot = -m2*g*l2*np.sin(theta2)+C1-C2
    
        return theta1_dot, theta2_dot, p1_dot, p2_dot
    
    def solve_polar(self):
        traj = integrate.solve_ivp(self.f,(0,t_end),(self.theta1,self.theta2,self.p1,self.p2),max_step = self.dt, atol = 1, method = 'RK45')
        return np.array((traj.y[0],traj.y[1],traj.y[2],traj.y[3],traj.t))
    
    def polar_to_cart(self,theta1,theta2):
        x1 = self.l1 * np.sin(theta1)
        y1 = -self.l1 * np.cos(theta1)
        x2 = x1 + self.l2 * np.sin(theta2)
        y2 = y1 - self.l2 * np.cos(theta2)
        return np.array((x1,y1,x2,y2))
    
    def solve_cart(self):
        traj_polar = self.solve_polar()
        pos = self.polar_to_cart(traj_polar[0],traj_polar[1]) #(theta1, theta2)
        return np.array((pos[0],pos[1],pos[2],pos[3],traj_polar[4]))
        
```

```python
def rotate(x, y, angle):
    rx = np.cos(angle)*x-np.sin(angle)*y
    ry = np.sin(angle)*x+np.cos(angle)*y
    return rx, ry
def cart_to_polar(x1,y1,x2,y2):
    theta1 = np.arctan2(y1,x1)
    theta2 = np.arctan2(y2,x2)
    return np.array((theta1,theta2))
def polar_to_cart(theta1,theta2):
    x1 = l1 * np.sin(theta1)
    y1 = -l1 * np.cos(theta1)
    x2 = x1 + l2 * np.sin(theta2)
    y2 = y1 - l2 * np.cos(theta2)
    return np.array((x1,y1,x2,y2))
```

```python
# Constants
l1 = 1
l2 = 1
m1 = 1
m2 = 1
dt = 0.002
t_end = 200
# IC's
# Low Energy
#theta1 = np.pi/3
#theta2 = 0
# High Energy
theta1 = 2*np.pi/3
theta2 = np.pi/3

pend_1 = DoublePendulum(theta1,theta2,l1,l2,m1,m2,dt,t_end)
traj_c = pend_1.solve_cart() #(x1,y1,x2,y2,t)
traj_p = pend_1.solve_polar() #(theta1,theta2,p1,p2,t)
```

```python
# Convert data to NN format
train_size = 1500
n_M = 50
train_count = 0
train_in = np.zeros((train_size,n_M+1,2))
train_out_theta  = np.zeros((train_size,2))
train_out_p  = np.zeros((train_size,2))

train_in_1 = np.zeros((train_size,n_M+1))
train_out_1_theta  = np.zeros((train_size))
train_out_1_p  = np.zeros((train_size))

for i in range(train_size): # Loop through trajectory
        init_t = int(np.floor(np.random.uniform(0,t_end/dt+1-(n_M+2))))
        # train for theta1
        train_in[train_count,:,0] = traj_p[0,init_t:init_t+n_M+1] # n-n_m to n points (ie. current n and all memories)
        train_out_theta[train_count,0] = traj_p[0,init_t+n_M+1] # n+1 point
        train_out_p[train_count,0] = traj_p[2,init_t+n_M+1] # n+1 point
        
        train_in_1[train_count,:] = traj_p[0,init_t:init_t+n_M+1] # n-n_m to n points (ie. current n and all memories)
        train_out_1_theta[train_count] = traj_p[0,init_t+n_M+1] # n+1 point
        train_out_1_p[train_count] = traj_p[2,init_t+n_M+1] # n+1 point
        
        # train for theta2
        train_in[train_count,:,1] = traj_p[1,init_t:init_t+n_M+1] # n-n_m to n points (ie. current n and all memories)
        train_out_theta[train_count,1] = traj_p[1,init_t+n_M+1] # n+1 point
        train_out_p[train_count,1] = traj_p[3,init_t+n_M+1] # n+1 point
        
        train_count += 1
```

```python
# MomentumRNN
# First, let's define a momentumRNN Cell, as a layer subclass # https://www.tensorflow.org/api_docs/python/tf/keras/layers/RNN
class MomentumRNNCell(tf.keras.layers.Layer):
    def __init__(self, units, **kwargs):
        self.units = units
        self.state_size = [units,units]
        super(MomentumRNNCell, self).__init__(**kwargs)
    
    def build(self, input_shape):
        self.kernel = self.add_weight(shape=(input_shape[-1], self.units),
                                      initializer='glorot_uniform',
                                      name='kernel')
        self.recurrent_kernel = self.add_weight(shape=(self.units, self.units),
                                      initializer='orthogonal',
                                      name='recurrent_kernel')
        self.built = True
        
    def call(self, inputs, states):
        stepsize = 0.002
        momentum = 1
        
        prev_h = states[0]
        prev_v = states[1]
        
        x = (tf.keras.backend.dot(inputs, self.kernel)) * stepsize
        v_out = prev_v * momentum + x
        h_out = tf.keras.activations.tanh(tf.keras.backend.dot(prev_h, self.recurrent_kernel) + v_out)
        
        return h_out, [h_out, v_out]

inputs = tf.keras.Input((n_M+1,2))

m_rnn, state_h, state_v = tf.keras.layers.LSTM(30,return_state=True,return_sequences=True)(inputs)
m_rnn, state_h, state_v = tf.keras.layers.LSTM(30,return_state=True,return_sequences=True)(m_rnn)
m_rnn, state_h, state_v = tf.keras.layers.LSTM(30,return_state=True)(m_rnn)

outputs = tf.keras.layers.Dense(2)(m_rnn)
m_rnn_model = tf.keras.Model(inputs,outputs)
m_rnn_model.summary(line_length = 200)
```

```python
tf.random.set_seed(0)
optimizer = tf.keras.optimizers.Adam(lr = .001)
m_rnn_model.compile(optimizer=optimizer,loss=tf.keras.losses.MeanSquaredError())
history_m_rnn  = m_rnn_model.fit(train_in.astype('float32'),train_out_theta.astype('float32'),epochs=20,batch_size=10,shuffle=True,verbose=1)

optimizer = tf.keras.optimizers.Adam(lr = .0001)
m_rnn_model.compile(optimizer=optimizer,loss=tf.keras.losses.MeanSquaredError())
history_m_rnn  = m_rnn_model.fit(train_in.astype('float32'),train_out_theta.astype('float32'),epochs=20,batch_size=10,shuffle=True,verbose=1)

optimizer = tf.keras.optimizers.Adam(lr = .00001)
m_rnn_model.compile(optimizer=optimizer,loss=tf.keras.losses.MeanSquaredError())
history_m_rnn  = m_rnn_model.fit(train_in.astype('float32'),train_out_theta.astype('float32'),epochs=20,batch_size=10,shuffle=True,verbose=1)
```
